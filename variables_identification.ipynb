# creating Titanic CSV data

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('https://www.openml.org/data/get_csv/16826755/phpMYEkMl')
```

```python
data
```

```python
data = data.replace('?', np.nan)
```

```python
def get_first_cabin(cabin):
  try:
    return cabin.split()[0]
  except:
    return np.nan
```

```python
data['cabin'] = data['cabin'].apply(get_first_cabin)
```

```python
data.to_csv('titanic.csv')
```
